```python
# importing the essential libraries we'll need for data handling, visualization and ML
import pandas as pd  # for working with dataframes easily
import numpy as np   # for numerical calculations
import matplotlib.pyplot as plt  # to create charts
import seaborn as sns  # for better and cleaner looking plots

# libraries for preprocessing and model building
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.linear_model import LinearRegression
# Import NearestNeighbors
from sklearn.neighbors import NearestNeighbors

# this line just makes sure the plots show up in the notebook
%matplotlib inline

# %%
# loading our dataset
df = pd.read_csv("/content/dataset.csv")  # make sure the file is uploaded
df.head()

# %%
# encoding categorical columns to prepare for modeling
df_encoded = df.copy()
label_enc = LabelEncoder()

for col in df_encoded.select_dtypes(include='object').columns:
    df_encoded[col] = label_enc.fit_transform(df_encoded[col])

# %%
# scaling the data to bring all values to similar range
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_encoded)

# %% [markdown]
#
# %%
# using K-Nearest Neighbors to recommend similar profiles
model_knn = NearestNeighbors(n_neighbors=5, algorithm='auto')
model_knn.fit(df_scaled)
```

```python
# selecting a random investor (you can change the index)
index = 10

# getting neighbors (similar investors)
distances, indices = model_knn.kneighbors([df_scaled[index]])

# displaying recommendations
print("Based on the investor at index", index, "we recommend:")
df.iloc[indices[0]]
```

```python
# Predicting Mutual Fund investment tendency as a proxy for expected return (since no 'Expected_Returns' column)

# Dropping target and irrelevant columns
features = df.drop(columns=['Mutual_Funds', 'Avenue', 'Investment_Avenues'])  # target + summary cols
target = df['Mutual_Funds']  # numeric proxy target

# Encoding and scaling
features_encoded = features.copy()
le = LabelEncoder()
for col in features_encoded.select_dtypes(include='object').columns:
    features_encoded[col] = le.fit_transform(features_encoded[col])

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_encoded)
```

```python
# Splitting data for training and testing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(
    features_scaled, target, test_size=0.2, random_state=42)

# Training the model
reg_model = LinearRegression()
reg_model.fit(X_train, y_train)

# Making predictions
y_pred = reg_model.predict(X_test)
```

```python
# Comparing actual and predicted mutual fund values
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(y_test[:50], y_pred[:50], color='purple')
plt.plot([0, max(y_test)], [0, max(y_test)], 'r--', label='Perfect prediction')
plt.title("Actual vs Predicted Mutual Fund Tendency")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()
plt.grid(True)
plt.show()
```
