```python
#importing allthe main libraries we need for classification tasks
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split #help in spliting dataset into tarin and test sets
from sklearn.preprocessing import LabelEncoder #for encoding categorical data
from sklearn.ensemble import RandomForestClassifier #a strong classififcation model(esemble method)
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix #for checking how good our model is
```

```python
import pandas as pd

#load csv file
df=pd.read_csv('/content/dataset.csv')
print("data loaded successfully")
df.head()
```

```python
#let's check the column names  and data types
print("columns in dataset:")
print(df.columns)

#overview of data types and missing values
df.info()
```

```python
#check missing values again in case you haven't done it in this new notebook
missing_values=df.isnull().sum()
print("missing values in each column:")
print(missing_values)
```

```python
from sklearn.preprocessing import LabelEncoder

cat_cols = df.select_dtypes(include = 'object').columns
print("categoricalcolumns:", cat_cols)

encoded_df = df.copy()
label_enc = LabelEncoder()

for col in cat_cols:
  if encoded_df[col].nunique() <= 2:
    encoded_df[col] = label_enc.fit_transform(encoded_df[col])
  else:
    encoded_df = pd.get_dummies(encoded_df, columns=[col], prefix=col)

print("categorical encof=ding completed")
```

```python
target_col = 'Avenue'

# Get the one-hot encoded columns for 'Avenue'
avenue_cols = [col for col in encoded_df.columns if col.startswith('Avenue_')]

# Drop the one-hot encoded columns related to 'Avenue' from features
x = encoded_df.drop(columns=avenue_cols)  # Changed from [target_col] to avenue_cols
y = df[target_col] #taking orignal version

#encode the target column
y_encoded = label_enc.fit_transform(y)
print("target classes:", label_enc.classes_)
```

```python
# Let's split our data into training and testing sets to train and evaluate the model properly

from sklearn.model_selection import train_test_split

# We’ll use 80% of the data for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(
    x, y_encoded,          # input features and encoded target, changed X to x
    test_size=0.2,         # 20% data for testing
    random_state=42,       # random_state helps us get the same split every time
    stratify=y_encoded     # ensures each class is represented proportionally in train & test
)

# Print the shape of each set to make sure everything is correct
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Time to build our model!
# I'm using Random Forest because it's a solid all-rounder for classification problems

from sklearn.ensemble import RandomForestClassifier

# Creating the model
model = RandomForestClassifier(random_state=42)

# Training the model using our training data
model.fit(X_train, y_train)

print("Model training complete ✅")
```

```python
# Now let's use our trained model to predict on the test set

y_pred = model.predict(X_test)

# Printing out the first few predictions to take a look
print("Sample predictions on test data:", y_pred[:10])
```

```python
# Let's evaluate how well our model is performing

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Accuracy tells us the overall percentage of correct predictions
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of the model:", round(accuracy * 100, 2), "%")

# Classification report gives detailed performance for each class (like Mutual Funds, Bonds, etc.)
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix helps us visualize how predictions compare to actual values
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Our goal is to classify if someone is Risk-Averse or Risk-Seeking
# Let's assume the 'Factor' column represents this (based on context)

# First, we'll take the same features (X) we used before
x_risk = x.copy()  # Just to keep things clean

# Target variable is 'Factor'
y_risk = df['Factor']

# Let's check unique values in the target column
print("Unique values in 'Factor':", y_risk.unique())
```

```python
# We'll encode the target so that the model understands it
# e.g., Risk-Averse → 0, Risk-Seeking → 1

from sklearn.preprocessing import LabelEncoder

le_risk = LabelEncoder()
y_risk_encoded = le_risk.fit_transform(y_risk)

# Just confirming how the encoding looks
print("Encoded labels:", set(zip(y_risk, y_risk_encoded)))
```

```python
from sklearn.model_selection import train_test_split

# Splitting data for training and testing the risk prediction model
# Removing 'stratify' since it causes issues with classes having only one sample
X_risk_train, X_risk_test, y_risk_train, y_risk_test = train_test_split(
    x_risk, y_risk_encoded,
    test_size=0.2,
    random_state=42
    # stratify=y_risk_encoded  # Removed this line
)

# Just checking the shape of the splits
print("X_risk_train:", X_risk_train.shape)
print("X_risk_test:", X_risk_test.shape)
print("y_risk_train:", y_risk_train.shape)
print("y_risk_test:", y_risk_test.shape)
```

```python
# Let's import the logistic regression model from sklearn
from sklearn.linear_model import LogisticRegression

# We'll also import metrics to evaluate the model
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Now let's initialize and train the logistic regression model on training data
risk_model = LogisticRegression()

# Fit the model to our training data
risk_model.fit(X_risk_train, y_risk_train)
```

```python
# After training, we'll use the model to make predictions on the test set
y_risk_pred = risk_model.predict(X_risk_test)

# Let's print predictions for quick check
print("Predicted Risk Labels:", y_risk_pred[:10])  # just showing first 10
```

```python
# Let's calculate accuracy to check how many predictions were correct
risk_accuracy = accuracy_score(y_risk_test, y_risk_pred)
print(f"Accuracy of Risk Prediction Model: {risk_accuracy:.2f}")
```

```python
# Confusion Matrix shows correct vs wrong predictions by category
print("Confusion Matrix:")
print(confusion_matrix(y_risk_test, y_risk_pred))

# Classification report gives precision, recall, f1-score for both classes
print("\nClassification Report:")
print(classification_report(y_risk_test, y_risk_pred))
```

```python
# Import LabelEncoder to convert duration categories into numeric labels
from sklearn.preprocessing import LabelEncoder

# Extracting features and target variable
# Find columns related to 'Duration' after one-hot encoding
duration_cols = [col for col in encoded_df.columns if col.startswith('Duration_')]

# Drop the one-hot encoded columns related to 'Duration' from features
x_duration = encoded_df.drop(columns=duration_cols, axis=1)
y_duration = df['Duration'] # Taking original version before encoding

# Encoding the Duration target variable
duration_encoder = LabelEncoder()
y_duration_encoded = duration_encoder.fit_transform(y_duration)

# Just confirming the encoded classes
print("Encoded Duration Classes:", list(duration_encoder.classes_))
```

```python
from sklearn.model_selection import train_test_split

# Splitting data (no stratify due to class imbalance)
X_dur_train, X_dur_test, y_dur_train, y_dur_test = train_test_split(
    x_duration, y_duration_encoded,
    test_size=0.2,
    random_state=42
)

# Check shapes of the datasets
print("X_dur_train:", X_dur_train.shape)
print("X_dur_test:", X_dur_test.shape)
print("y_dur_train:", y_dur_train.shape)
print("y_dur_test:", y_dur_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Using Random Forest Classifier for multiclass classification
duration_model = RandomForestClassifier(random_state=42)

# Training the model
duration_model.fit(X_dur_train, y_dur_train)
```

```python
# Predicting the duration class using trained model
y_dur_pred = duration_model.predict(X_dur_test)

# Checking first few predictions
print("Predicted Duration (Encoded):", y_dur_pred[:10])
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Accuracy of duration model
duration_accuracy = accuracy_score(y_dur_test, y_dur_pred)
print(f"Accuracy of Duration Prediction Model: {duration_accuracy:.2f}")

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_dur_test, y_dur_pred))

# Classification Report
print("\nClassification Report:")
print(classification_report(y_dur_test, y_dur_pred))
```

```python
# Converting numeric predictions and actual labels back to original class names
y_dur_pred_labels = duration_encoder.inverse_transform(y_dur_pred)
y_dur_test_labels = duration_encoder.inverse_transform(y_dur_test)

# Display first few decoded values
print("Actual Duration Labels:", y_dur_test_labels[:10])
print("Predicted Duration Labels:", y_dur_pred_labels[:10])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

# Create confusion matrix
cm = confusion_matrix(y_dur_test_labels, y_dur_pred_labels, labels=duration_encoder.classes_)

# Plot using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='YlGnBu',
            xticklabels=duration_encoder.classes_,
            yticklabels=duration_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Investment Duration Classification')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

# Perform 5-fold cross-validation to see consistency
cv_scores = cross_val_score(duration_model, x_duration, y_duration_encoded, cv=5)

# Display results
print("Cross-validation scores:", cv_scores)
print("Average CV accuracy:", round(cv_scores.mean(), 2))
```

```python
# Check which features are influencing the prediction the most
importances = duration_model.feature_importances_
features = x_duration.columns

# Sorting top 10 features
feat_imp = pd.Series(importances, index=features).sort_values(ascending=False)[:10]

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=feat_imp, y=feat_imp.index, palette="viridis")
plt.title('Top 10 Important Features for Duration Prediction')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()
```
