```python
# importing all the required libraries here
import pandas as pd   # for handling and playing with tabular data
import numpy as np    # helpful for numerical operations
import matplotlib.pyplot as plt  # for making visualizations
import seaborn as sns  # gives better-looking plots than default

# now reading the CSV file (already uploaded to the session or Drive)
df = pd.read_csv("dataset.csv")

# just checking the first few rows to get an idea of what data looks like
df.head()
```

```python
# checking if there are any missing values in each column
df.isnull().sum()
```

```python
df.replace(['?','N\A','na','Na','','Missing'],np.nan,inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(),cmap='viridis',cbar=False)
plt.title('missing values Heatmap in Dataset')
plt.show()

```

```python
# checking the data types of each column
df.dtypes
```

```python
num_cols=df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
  if df[col].isnull().sum()>0:
    df[col].fillna(df[col].mean(),inplace=True)
```

```python
cat_cols=df.select_dtypes(include=['object']).columns

for col in cat_cols:
  if df[col].isnull().sum()>0:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# verifying again if all missing values are handled
df.isnull().sum()
```

```python
#first identify which columns in our dataset are categorical so we can encode them later
cat_cols = df.select_dtypes(include = 'object').columns
print("categorical column in dataset:",cat_cols)
```

```python
#for each categorical columns let's check how many unique values it has(important to choose encode types)
for col in cat_cols:
  print(f"{col}: {df[col].nunique} unique values")
```

```python
#now apply encoding : using label enoding if unique values are few,else we will use one-hot encoding forcolumn with more than two values
from sklearn.preprocessing import LabelEncoder
#making a copy of the orignal df to safety perform encoding
encoded_df = df.copy()
label_enc = LabelEncoder()

#Loop through categorical columns and encode smartly
for col in cat_cols:
  if encoded_df[col].nunique() <= 2:
    #use label encoding when only 2 categories (like gender)
    encoded_df[col] = label_enc.fit_transform(encoded_df[col])

  else:
    # if more than 2 categories,use one hot encoding
    encoded_df = pd.get_dummies(encoded_df, columns=[col], prefix=col)

#after encoding is done,let's just check how the new dataaframe look like
print("encoding completed.here's a preview of the new dataframe:")
encoded_df.head()

```

```python
#firsst,let's identify numerical columns from the dataset
num_cols = encoded_df.select_dtypes(include=['int64', 'float64']).columns
print("numerical columns in dataset:",num_cols)

#now let's describe them to central tendency,spread,etc.
print("\ndescriptive static of numerical coluumns:")
encoded_df[num_cols].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#let's plot the distribution of each numerical column using histograms and KDE

for col in num_cols:
  plt.figure(figsize=(8,4)) #setting size of the plot
  sns.histplot(encoded_df[col], kde=True, bins=30, color='skyblue') #KDE=True shows distribution curve
  plt.title(f'distribution of {col}' ,fontsize=14)
  plt.xlabel(col)
  plt.ylabel('frequency')
  plt.grid(True)
  plt.tight_layout()
  plt.show()
```

```python
#calculate coorelation matrix for numerical columns
coorelation_matrix = encoded_df[num_cols].corr()

#set the size of heatmap
plt.figure(figsize=(12,8))

#createe heatmap using seaborn
sns.heatmap(coorelation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

#set title and show the plot
plt.title('coorelation Heatmap of numerical features', fontsize=16)
plt.tight_layout()
plt.show()
```
